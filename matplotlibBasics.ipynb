*Simple Line Plot*

```python
x=np.linspace(0,10,50)
y=np.sin(x)
plt.plot(x,y)
plt.title("Sine wave")
plt.xlabel("X axis")
plt.ylabel("Y = sin(x)")
plt.show()
```
