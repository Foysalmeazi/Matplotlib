*Simple Line Plot*

```python
x=np.linspace(0,10,50)
y=np.sin(x)
plt.plot(x,y)
plt.title("Sine wave")
plt.xlabel("X axis")
plt.ylabel("Y = sin(x)")
plt.show()
```

*Customied Plot*

```python
plt.figure(figsize=(7,4))
plt.plot(x,y, color="red", linestyle="--", label="Sine wave")
plt.title("Customized Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.grid()
plt.show()
```
