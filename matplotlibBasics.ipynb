*Simple Line Plot*

```python
x=np.linspace(0,10,50)
y=np.sin(x)
plt.plot(x,y)
plt.title("Sine wave")
plt.xlabel("X axis")
plt.ylabel("Y = sin(x)")
plt.show()
```

*Customied Plot*

```python
plt.figure(figsize=(7,4))
plt.plot(x,y, color="red", linestyle="--", label="Sine wave")
plt.title("Customized Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

A bar chart is used to compare different categories or groups side-by-side.Think of it as a visual tool to answer the question: "Which specific group is the biggest, smallest, or most popular?"

*Bar Chart*

```python
students = ['Foysal', 'Hasib', 'Rakib', 'Rasel', 'Raju']
scores = [80, 78, 65, 60, 58]

plt.bar(students, scores, color='blue')
plt.title("Student Scores - Vertical Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.ylim(0, 100)
plt.show()
```

```python
students = ['Foysal', 'Hasib', 'Rakib', 'Rasel', 'Raju']
scores = [80, 78, 65, 60, 58]

plt.bar(students, scores, color='orange')
plt.title("Student Scores - Horizontal Bar Chart")
plt.xlabel("Scores")
plt.ylabel("Students")
plt.ylim(0, 100)
plt.show()
```

*Scatter Plot*. A scatter plot is used to see if two different things are connected or related to each other

```python
np.random.seed(0)

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green', alpha=0.9)
plt.title("Random Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

Histogram normally use for frequency distribution (how many times a data point falls into that specific range on the vertical Y-axis)

*Histogram (Normal Data Distribution)*

```python
np.random.seed(0)
data = np.random.randn(500)

plt.hist(data, bins=20, color='purple', edgecolor='black', alpha=0.7)
plt.title("Histogram of Random Values")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

*Histogram (Right Skewed Data Distribution)*

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.exponential(scale=2, size=500)

plt.hist(data, bins=30, color='purple', edgecolor='black', alpha=0.7)
plt.title("Right-Skewed Data Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

Pie Chart

```python
labels = ['Python', 'Java', 'JavaScript', 'C++', 'R']
popularity = [30, 25, 20, 15, 10]

plt.pie(popularity, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Programming Language Popularity")
```

Multiple Plots (Sub Plots)

```python
x = np.linspace(0,20,120)
y=np.sin(x)
z=np.cos(x)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(x, y, 'r--', label='Sine Wave')
plt.title("Sine Waves")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()

plt.subplot(1,2,2)
plt.plot(x, z, 'b-', label='Cosine Wave')
plt.title("Cosine Waves")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.suptitle("Sine and Cosine Waves")
```
