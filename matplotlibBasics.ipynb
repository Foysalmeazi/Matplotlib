*Simple Line Plot*

```python
x=np.linspace(0,10,50)
y=np.sin(x)
plt.plot(x,y)
plt.title("Sine wave")
plt.xlabel("X axis")
plt.ylabel("Y = sin(x)")
plt.show()
```

*Customied Plot*

```python
plt.figure(figsize=(7,4))
plt.plot(x,y, color="red", linestyle="--", label="Sine wave")
plt.title("Customized Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

*Bar Chart*

```python
students = ['Foysal', 'Hasib', 'Rakib', 'Rasel', 'Raju']
scores = [80, 78, 65, 60, 58]

plt.bar(students, scores, color='blue')
plt.title("Student Scores - Vertical Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.ylim(0, 100)
plt.show()
```
